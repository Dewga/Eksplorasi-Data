### Setelah cek value terbesarnya berapa lanjut cek outliernya


```python
q1 = df["fare"].quantile(0.25)
q3 = df["fare"].quantile(0.75)
iqr = q3 - q1

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

lower, upper
```

```python
fare_outliers = df[df["fare"] > upper]
fare_outliers[["fare","sex","pclass","survived"]].sort_values("fare", ascending=False).head(10)
```
